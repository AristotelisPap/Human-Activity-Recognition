**Question c -> i**

Some common types of time-domain features that are usually used in time series classification include:

1. **Mean** - Average value of the signal over the time window 
2. **Median** - Median value of the signal over the time window 
3. **Standard deviation** - Measure of the spreadness of the signal over the time window
4. **Variance** - Squared standard deviation
5. **Root Mean Square (RMS)** - The quadratic mean value of the signal over the time window
6. **Min** - The minimum value of the signal over the time window
7. **Max** - The maximum value of the signal over the time window
8. **Interquartile range** - Measure of the statistical dispersion, being equal to the difference between the 3rd quartile and the 1st quarile of the signal over the time window
9. **Zero crossing rate** - Rate of sign-changes along a signal
10. **Mean crossing rate** - Rate of signal changes from below average to above average or vice versa
11. **Pairwise Correlation** - Correlation between two axes of each sensor and different sensors
12. **Spectral Entropy** - Measure of the distribution of frequency components
13. **Averaged Derivatives** - The mean value of the first order derivatives of the signal over the time window
14. **Skewness** - The degree of asymmetry of the sensor signal distribution
15. **Kurtosis** - The degree of peakedness of the sensor signal distribution


**References**

**[1]** M. Zhang, A. A. Sawchuk, "A Feature Selection-Based Framework for Human Activity Recognition Using Wearable Multimodal Sensors", *International Conference on Body Area Networks (BodyNets)*, Beijing, China, November 2011.

```python
# Question c -> ii

import numpy as np

# In this cell, I will first merge all the dataframes (both training and test) of each activity
# into a single dataframe in order to be able to convert it afterwards into the form that is requested 
# from the question.

bending1 = [bending1_test, bending1_train]
bending1 = pd.concat(bending1)
bending2 = [bending2_test, bending2_train]
bending2 = pd.concat(bending2)
cycling = [cycling_test, cycling_train]
cycling = pd.concat(cycling)
lying = [lying_test, lying_train]
lying = pd.concat(lying)
sitting = [sitting_test, sitting_train]
sitting = pd.concat(sitting)
standing = [standing_test, standing_train]
standing = pd.concat(standing)
walking = [walking_test, walking_train]
walking = pd.concat(walking)

# Now, I will unify all them into a sinle dataframe!
total_data = [bending1, bending2, cycling, lying, sitting, standing, walking]
total_data = pd.concat(total_data)


# Create empty lists
min_list = []
max_list = []
mean_list = []
median_list = []
std_list = []
first_quart_list = []
third_quart_list = []

# Save the requested features inside the lists
for i in range(0, 42240, 480):
    min_list.append(total_data.iloc[i:480+i, 1:7].min(axis = 0))  
    max_list.append(total_data.iloc[i:480+i, 1:7].max(axis = 0))
    mean_list.append(total_data.iloc[i:480+i, 1:7].mean(axis = 0))
    median_list.append(total_data.iloc[i:480+i, 1:7].median(axis = 0))
    std_list.append(total_data.iloc[i:480+i, 1:7].std(axis = 0))
    first_quart_list.append(total_data.iloc[i:480+i, 1:7].quantile(q=0.25, axis=0, interpolation='midpoint'))
    third_quart_list.append(total_data.iloc[i:480+i, 1:7].quantile(q=0.75, axis=0, interpolation='midpoint'))
    
    
# Convert the lists into numpy arrays    
minarray = np.asarray(min_list)
maxarray = np.asarray(max_list)
meanarray = np.asarray(mean_list)
medianarray = np.asarray(median_list)
stdarray = np.asarray(std_list)
first_quartarray = np.asarray(first_quart_list)
third_quartarray = np.asarray(third_quart_list)


# Convert the numpy arrays into dataframes and name their columns
mindf = pd.DataFrame(data=minarray,index=None,columns=None)
mindf.columns = ["min1", "min2", "min3", "min4", "min5", "min6"]
maxdf = pd.DataFrame(data=maxarray,index=None,columns=None)
maxdf.columns = ["max1", "max2", "max3", "max4", "max5", "max6"]
meandf = pd.DataFrame(data=meanarray,index=None,columns=None)
meandf.columns = ["mean1", "mean2", "mean3", "mean4", "mean5", "mean6"]
mediandf = pd.DataFrame(data=medianarray,index=None,columns=None)
mediandf.columns = ["median1", "median2", "median3", "median4", "median5", "median6"]
stddf = pd.DataFrame(data=stdarray,index=None,columns=None)
stddf.columns = ["std1", "std2", "std3", "std4", "std5", "std6"]
first_quartdf = pd.DataFrame(data=first_quartarray,index=None,columns=None)
first_quartdf.columns = ["first_quart1", "first_quart2", "first_quart3",
                         "first_quart4", "first_quart5", "first_quart6"]
third_quartdf = pd.DataFrame(data=third_quartarray,index=None,columns=None)
third_quartdf.columns = ["third_quart1", "third_quart2", "third_quart3",
                         "third_quart4", "third_quart5", "third_quart6"]


# Merge the columns of the dataframes
merge = [mindf, maxdf, meandf, mediandf, stddf, first_quartdf, third_quartdf]
new_dataset = pd.concat(merge, axis=1)
# Reorder the columns in order to appear as requested from the question
cols = ['min1', 'max1', 'mean1', 'median1', 'std1', 'first_quart1', 'third_quart1',
       'min2', 'max2', 'mean2', 'median2', 'std2', 'first_quart2', 'third_quart2',
       'min3', 'max3', 'mean3', 'median3', 'std3', 'first_quart3', 'third_quart3',
       'min4', 'max4', 'mean4', 'median4', 'std4', 'first_quart4', 'third_quart4',
       'min5', 'max5', 'mean5', 'median5', 'std5', 'first_quart5', 'third_quart5',
       'min6', 'max6', 'mean6', 'median6', 'std6', 'first_quart6', 'third_quart6']
new_dataset = new_dataset[cols]
new_dataset
```

As it is also mentioned and illustrated in the code of the next cell, the **RIGHT** way to perform cross-validation
is to incorporate the feature selection process inside the loop of cross-validation.

The **WRONG** way is to select the features first and then cross-validate in order to get the best value of the parameter l. This approach does not take into account the fact that the held-out fold will already have seen all the labels if the feature selection happens before the cross-validation!!

```python
# Question d -> iii (Continue)

# Now, I need to pick the best l and concurrently do feature selection
# by using the 5-fold cross validation technique.

from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

from sklearn.model_selection import KFold
# Due to imbalanced datasets, I had to use Stratified k-fold CV !!
from sklearn.model_selection import StratifiedKFold

# Define my classifier to be Logistic Regression
model = LogisticRegression()

# I will use 5-fold cross validation in order to calculate both the best number
# features that we should select as well as the best value for l. 

# Initialize a list where I am going to save the average CV errors for the
# 20 different values of l times the 42 different possible number of features
# that we should select!!!
ave_CV_errors = []

for l in range(1, 21):
    X = training_sets['training_set_l' + str(l)].iloc[:, 0:42]
    y = training_sets['training_set_l' + str(l)].iloc[:, 42:43]
    for num_of_feat in range(1, 43):   
        skf = StratifiedKFold(n_splits=5) # I will use 5-fold cross validation!
        skf.get_n_splits(X)
        # Initialize a list in order to get the errors from 5-fold CV
        list_5fold = []
        for train_index, test_index in skf.split(X,y):
            X = training_sets['training_set_l' + str(l)].iloc[:, 0:42]
            y = training_sets['training_set_l' + str(l)].iloc[:, 42:43]
            y = np.ravel(y) # In order to stop some warnings having to do with the definition of the shape
            X_train, X_test = X.values[train_index], X.values[test_index]
            # Convert them into Dataframes in order to use the iloc command afterwards
            X_train = pd.DataFrame(data=X_train,index=None,columns=cols)
            X_test = pd.DataFrame(data=X_test,index=None,columns=cols)
            y_train, y_test = y[train_index], y[test_index]
            rfe = RFE(model, num_of_feat) 
            # I will apply feature selection inside the cross-validation!
            # Moreover, I will apply it in the k-1 folds and I will not let the held-out fold
            # to see the labels.
            rfe = rfe.fit(X_train, y_train)
            # Now, I need to choose those features that RFE suggested.
            sel_features = [] # Create an empty list
            for i in range(0, 42):
                if (rfe.ranking_[i] == 1):
                    sel_features.append(i) # Save the indexes of the features that RFE marked as important
                    # into a list.
            # Keep only these features
            X_train = X_train.iloc[:, sel_features]
            X_test = X_test.iloc[:, sel_features] # Note that X_test did not participate in the
            # feature selection process!!!
            # Now, I will fit my logistic regression model in the k-1 folds!
            model.fit(X_train, y_train)
            list_5fold.append(1 - model.score(X_test, y_test))
        # Calculate the average error of list_5fold and save it in ave_CV_errors list
        ave_CV_errors.append(sum(list_5fold) / len(list_5fold))

# Convert the ave_CV_errors list into a numpy array
Arrray = np.asarray(ave_CV_errors)
Arrray = Arrray.reshape(20,42) # Indexes are (l, num_of_feat)
# Take the indexes of the minimum element of the Arrray
ind = np.unravel_index(np.argmin(Arrray, axis=None), Arrray.shape)
    
        
# The list ave_CV_errors contains the average CV errors obtained by the 20 different values of l
# and the 42 possible numbers of features that RFE should select.
# In order to pick the best l, we are going to pick the lowest value out of this list!!
print("The minimum CV error happens for : l =", ind[0] + 1)
print("The best number of features selected is : number of features =", ind[1] + 1)
print("The CV error for l = 1 is: ", min(ave_CV_errors))
```

After cross-validation, I got that the best value for the parameter l is **l = 1**.

Moreover, I got that given l=1, the best number of features that should be selected is **number of features = 4**.

```python
# Question 1d -> iv

# Since we picked the best parameter to be l=1, we first have to re-train our model using the
# whole training data and re-applying the feature selection process!

l = ind[0] + 1 # This was the one that was calculated from CV

X_train = training_sets['training_set_l' + str(l)].iloc[:, 0:42]
y_train = training_sets['training_set_l' + str(l)].iloc[:, 42:43]
y_train = np.ravel(y_train) # In order to stop some warnings having to do with the definition of the shape

model = LogisticRegression()
rfe = RFE(model, ind[1] + 1) 
rfe = rfe.fit(X_train, y_train)

# Now, I need to choose those features that RFE suggested.
sel_features = [] # Create an empty list
for i in range(0, 42):
    if (rfe.ranking_[i] == 1):
        sel_features.append(i) # Save the indexes of the features that RFE marked as important
        # into a list.
# Keep only these features
X_train = X_train.iloc[:, sel_features]
# Now, I will fit my logistic regression model 
model.fit(X_train, y_train)
y_pred = model.predict(X_train)

# Report the confusion matrix on TRAINING data
from sklearn import metrics
print("The confusion matrix is:\n")
print(metrics.confusion_matrix(y_train, y_pred))
```

### **Question d -> vi**

Both the CV accuracy and the accuracy on the test set are very high, meaning that the two classes (**class 1: 'Bending'**, **class 2: 'All other Activities'**) are very well separated.

In the literature, it is known that when the classes are well separated, then the estimates of the Logistic Regression coefficients become unstable, i.e. they approach infinity. This problem can be easily seen from the form of the logit function:

log(P/(1-P)) = aX + b

When the classes are very well separated, P is very close to 1 and thus the parameters approach infinity!!
